# Now we have to divide data in Dependent and Independent variable

# We see Date column is very usefull for prediction but for simplicity we have to remove it because date formate is not proper

# Now we have to predict open price, so this column is dependenet varaible because it depend on High, Low, Close, Last, Turnover etc

```python
# Let's select our features
```

```python
# We just remove the date column because it is not usefull here
x=data.loc[:,'High':'Turnover (Lacs)'].values
```

```python
data.loc[:,'High':'Turnover (Lacs)']
```

```python
x
```

```python
type(x)
```

```python
y=data.loc[:,'Open'].values
```

```python
y
```

```python
type(y)
```

# Now we have to split data in training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

# Let's fit our LinearRegression model

```python
LR=LinearRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
LR.score(x_test,y_test)
```
